```python
import matplotlib.pyplot as plt
import dicom
import glob
import numpy as np

path = '/media/carloalberto/DATASETS/DICOM/00cba091fa4ad62cc3200a657aeb957e/'

def dicom_to_image(filename):
    dcm = dicom.read_file(filename)
    img = dcm.pixel_array
    img[img == -2000] = 0
    return img
```

```python
files = glob.glob('/media/carloalberto/DATASETS/DICOM/*/*.dcm')

f, plots = plt.subplots(4, 5, sharex='col', sharey='row', figsize=(10, 8))
for i in range(20):
    plots[i // 5, i % 5].axis('off')
    plots[i // 5, i % 5].imshow(dicom_to_image(np.random.choice(files)), cmap=plt.cm.bone)
    
```
